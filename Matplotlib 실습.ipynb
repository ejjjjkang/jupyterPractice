Matplotlib : 주로 2차원의 데이터를 시각화하기 위한 서드파티 패키지
- 동작하는 운영체제를 가리지 않음
- Matlab과 유사한 사용자 인터페이스를 가짐
- 그림 요소의 상세한 서식 설정 가능
- 다양한 출력 형식으로 저장 가능

```python
import matplotlib.pyplot as plt

#Matplot 그래프 출력시 show() 함수
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

dat = [0,1]
ax.plot(dat)
plt.show()
```

- 그래프 작성 전 피겨 오브젝트(피겨)와 서브 플롯 작성 필요

```python
#피겨 생성
fig = plt.figure()

#피겨 안에 서브 플롯을 3개 배치한다
ax1 = fig.add_subplot(221) #2행 2열 1번
ax2 = fig.add_subplot(222) #2행 2열 2번
ax3 = fig.add_subplot(223) #2행 2열 3번

for i, ax in enumerate([ax1, ax2, ax3],start = 1):
    txt = 'ax{0}\n(22{0})'.format(i)
    ax.text(0.2, 0.4, txt, fontsize =24)

plt.show()
```

```python
#피겨 오브젝트 작성과 서브 플롯 배치를 동시에 실행
fig,axes = plt.subplots(2,2)
print(type(axes), axes)
plt.show()
```

```python
#행렬의 위치를 지정하여 서브 플롯을 참조할 수 있음(행렬의 지정은 0부터 시작)

fig,axes = plt.subplots(2,2)
axes[0,1].set_title('Subplot 0-1')
plt.show()
```
